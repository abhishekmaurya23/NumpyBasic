To print the data types of both variables, you can use the `type()` function. Here's an example code snippet:


Using the `type()` function, we print the data types of both variables. The data type of `list_` is `<class 'list'>`, indicating it is a Python list. The data type of `array_list` is `<class 'numpy.ndarray'>`, indicating it is a NumPy array.


In the above code, we have the variables `list_` and `array_list` with different data types in their elements.

Similarly, we use another for loop to iterate through the elements of `array_list` and print the data type for each element.

The output displays the data type of each element in both `list_` and `array_list`.

Using loops, we iterate over the elements of both `list_` and `array_list` and print the data type of each element. In `list_`, all elements are of the `int` data type, while in `array_list`, all elements are of the NumPy `int64` data type.

Finally, we print the `zeros_matrix`, which displays the 3x3 matrix of zeros.
